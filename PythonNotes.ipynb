```python
class Car:
    """
    The Car class represents a car with a make, model, and year.
    It provides methods to display information and update the car's mileage.
    """

    def __init__(self, make, model, year):
        """
        Initialize a new Car instance.

        Parameters:
        make (str): The manufacturer of the car (e.g., 'Toyota').
        model (str): The model of the car (e.g., 'Corolla').
        year (int): The year the car was made (e.g., 2020).
        """
        self.make = make      # Store the car's manufacturer
        self.model = model    # Store the car's model
        self.year = year      # Store the car's production year
        self.mileage = 0      # Initialize mileage to zero

    def display_info(self):
        """
        Print the car's information in a readable format.
        """
        print(f"{self.year} {self.make} {self.model} - Mileage: {self.mileage} miles")

    def drive(self, miles):
        """
        Simulate driving the car by increasing its mileage.

        Parameters:
        miles (int): The number of miles driven.

        Raises:
        ValueError: If miles is negative.
        """
        if miles < 0:
            raise ValueError("Miles driven cannot be negative.")
        self.mileage += miles  # Add miles to the car's mileage

# Example usage:
my_car = Car("Toyota", "Corolla", 2020)  # Create a new Car object
my_car.display_info()                    # Display car info
my_car.drive(150)                        # Drive the car 150 miles
my_car.display_info()                    # Display updated info
```

```python
import csv
from collections import defaultdict, Counter

# Path to the input CSV file containing president data, adjust as needed
input_file = '../data/presidents.csv'

# Dictionaries to store party affiliation and president names
parties = {}      # Maps president name to party
presidents = {}   # Maps president number to name

# Open the CSV file for reading
with open(input_file, newline='') as prez_file:
    reader = csv.reader(prez_file)
    next(reader)  # Skip the header row
    for row in reader:
        pres_num = row[0]           # President number (as string)
        name     = row[1].strip()   # President name (remove extra spaces)
        party    = row[5].strip()   # Party affiliation (remove extra spaces)
        parties[name] = party       # Store party by president name
        presidents[pres_num] = name # Store name by president number

# Example lookups
print(presidents.get('13'))               # Get president with number 13
print(presidents.get('16'))               # Get president with number 16
print(presidents.get('48', "TBD"))        # Get president 48 or "TBD" if not found

# Loop through all presidents and print their number and name
for num, name in presidents.items():
    print(f"President number {num} was {name}")

# Count how many presidents belonged to each party
party_count = Counter(parties.values())
print("Party counts:", party_count)

# Save selected columns (number, name, party) to a new file
with open("out.txt", "w") as outfile:
    for num, name in presidents.items():
        outline = f"{num},{name},{parties[name]}\n"
        outfile.write(outline)
```

```python
# More Intermediate Python Fundamentals

# 1. Nested List Comprehensions
matrix = [[i * j for j in range(3)] for i in range(3)]
print("Matrix:", matrix)

# 2. Using any() and all()
bools = [True, False, True]
print("Any true?", any(bools))
print("All true?", all(bools))

# 3. Set operations
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print("Union:", set_a | set_b)
print("Intersection:", set_a & set_b)
print("Difference:", set_a - set_b)

# 4. Dictionary .get() and .setdefault()
d = {"x": 1}
print("Get x:", d.get("x"))
print("Get y (default 0):", d.get("y", 0))
d.setdefault("y", 2)
print("After setdefault:", d)

# 5. Sorting with key functions
words = ["banana", "apple", "cherry"]
words_sorted = sorted(words, key=len)
print("Sorted by length:", words_sorted)

# 6. Enumerate with custom start
for idx, val in enumerate(["a", "b", "c"], start=1):
    print(f"{idx}: {val}")

# 7. List flattening
nested = [[1, 2], [3, 4], [5]]
flat = [item for sublist in nested for item in sublist]
print("Flattened:", flat)

# 8. Using zip with multiple lists
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]
for x, y, z in zip(a, b, c):
    print(x, y, z)

# 9. Reversed iteration
for val in reversed([1, 2, 3]):
    print("Reversed:", val)

# 10. List slicing with steps
nums = list(range(10))
print("Every other:", nums[::2])
```

```python

# Python Data Science Fundamentals
import numpy as np

# 1. Numpy arrays
arr = np.array([1, 2, 3, 4, 5]) #np is imported as numpy
print("Numpy array:", arr) #arr = np.array
print("Mean:", np.mean(arr)) #mean of array
print("Standard deviation:", np.std(arr)) #standard deviation of array

# 2. Pandas DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
print("Describe:\n", df.describe())

# 3. Data selection and filtering
print("Ages > 28:\n", df[df['Age'] > 28]) #filtering ages greater than 28

# 4. Grouping and aggregation(any type of math)
df['Gender'] = ['F', 'M', 'M'] 
grouped = df.groupby('Gender')['Age'].mean()
print("Mean age by gender:\n", grouped)

# 5. Handling missing data
df.loc[1, 'Age'] = np.nan
print("With missing value:\n", df)
print("Fill missing:\n", df.fillna(df['Age'].mean()))

# 6. Visualization (basic)
import matplotlib.pyplot as plt
df['Age'].fillna(df['Age'].mean()).plot(kind='hist', title='Age Distribution')
plt.xlabel('Age')
plt.show()
```
