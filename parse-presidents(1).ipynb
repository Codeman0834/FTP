```python
import csv
from collections import Counter

input_file = '../data/presidents.csv'

parties = {}
presidents = {}

with open(input_file, newline='') as prez_file:
    reader = csv.reader(prez_file)
    next(reader)  # skip header row
    for row in reader:
        pres_num = row[0]
        name     = row[1].strip()
        party    = row[5].strip()
        parties[name] = party
        presidents[pres_num] = name

# Examples
print(presidents.get('13'))
print(presidents.get('16'))
print(presidents.get('48', "TBD"))

# Loop
for num, name in presidents.items():
    print(f"President number {num} was {name}")

# Party counts
party_count = Counter(parties.values())
print("Party counts:", party_count)

# Save a subset of columns to a file
with open("out.txt", "w") as outfile:
    for num, name in presidents.items():
        outline = f"{num},{name},{parties[name]}\n"
        outfile.write(outline)
```
